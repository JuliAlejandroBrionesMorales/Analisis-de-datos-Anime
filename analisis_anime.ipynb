```python
# Generamos una función para reparar nulos en el dataframe. 
def reparar_nulos(df, col):
    if df[col].dtype == 'int64' or df[col].dtype == 'float64': # Si los de la columna es del tipo númerica, los valores nulos de esa columna seran reemplazados por la media
        df[col] = df[col].fillna(df[col].mean())
    else:
        df[col] = df[col].fillna(df[col].mode()[0]) # Si los de la columna no son del tipo númerica,  los valores nulos de esa columna seran reemplazados por la moda
    return df 

for col in df.columns:
    reparar_nulos(df, col)
```

### 1. ¿En función de lo que dura su episodio son mas populares los animes cortos o largos?  Considerar solo para series de TV.

```python
# Esto sirve para poder imprimir todas las columnas
pd.set_option('display.max_columns', None) 
```

```python
# Imprimimos la cantidad de valores que tenemos en la columna "type"

df['type'].value_counts()
```

```python
# Creamos una dataFrame "dftv" donde solo se muestre serie de TV dentro de la columna "type"
dftv = df[df['type'] == 'TV'] #solo series de tv
```

```python
# Imprimimos el nuevo dftv comprobar que se ha haya llevado el filtrado
dftv
```

```python
#  Una vez seleccionado las series de TV mediante el nuevo dataframe, tenemos que comparar la duración. Para ello, nos fijamos en los valores únicos de la columna "duration"
dftv['duration'].unique()
```

```python
# Eliminamos el texto después del numero entero para quedarnos solo con el número
dftv['duration'] = dftv['duration'].str.replace(' min. per ep.', '')
dftv['duration'] = dftv['duration'].str.replace(' min per ep.', '')
dftv['duration'] = dftv['duration'].str.replace(' min.', '')
dftv['duration'] = dftv['duration'].str.replace('  min per ep', '')
dftv['duration'] = dftv['duration'].str.replace('per ep', '')
dftv['duration'] = dftv['duration'].str.replace(' min', '')
dftv['duration'] = dftv['duration'].str.replace(' min.', '')
```

```python
# Filtramos todas las filas donde la columna "duration" no contiene la cadena "sec" 
dftv = dftv[~dftv['duration'].str.contains('sec')]
dftv = dftv[~dftv['duration'].str.contains('sec.')]
dftv = dftv[~dftv['duration'].str.contains('sec per ep')]
dftv = dftv[~dftv['duration'].str.contains('sec. per ep')]
```

```python
# Filtramos todas las filas donde la columna "duration" no contiene la cadena "Unknown" 
dftv = dftv[~dftv['duration'].str.contains('Unknown')]
```

```python
# Imprimimos la columna "duration" para ver los valores únicos luego de los cambios anteriores 
dftv['duration'].unique()
```

```python
# Convertimos los valores de la columna "duration" a valores numéricos
dftv['duration'] = pd.to_numeric(dftv['duration'])
```

```python
# Imprimimos la columna "duration" para ver que todos los valores son numéricos
dftv['duration']
```

```python
# Como queremos comprobar si los episodios son cortos o largos, con "describe" imprimimos las estadísticas de los valores de la columna "duration"
dftv['duration'].describe()
```

```python
# Vemos los rangos intercuartilicos
sns.boxplot(dftv['duration'])
```

```python
# Como resultado de las estadísticas visto anteriormente, como el 50% de las series de TV son menores de 24 minutos,creamos dos grupos, las series de tv que duran mas de 24 minutos y los que duran menos 
dftv['duration'] = np.where(dftv['duration'] > 24, 'Largo', 'Corto') # si la duración es más de 24 min, son las series de tv son "largos", si no son "cortos"

dftv['duration'].value_counts() # Contamos la cantidad de series de tv cortas y largas
```

```python
# Ahora comprobamos con nuestra columna "score" si los animes cortos o largos son mas populares usando pandas 
dftv.groupby('duration')['score'].mean().round(2)
```

```python
# Ahora comprobamos con nuestra columna "popularity" si los animes cortos o largos son mas populares usando pandas 
dftv.groupby('duration')['popularity'].mean().round(2)
```

#### Conclusión:Si tenemos en cuenta el "score" para determinar si las series de tv largos o cortas son más populares, nos sale  como resultado que son muy parecidas en cuanto a popularidad. Por otro lado, si tenemos en cuenta "popularity" las series de tv largos son más populares. Para una mayor precisión en la conclusión, será necesario profundizar más estadísticamente.
